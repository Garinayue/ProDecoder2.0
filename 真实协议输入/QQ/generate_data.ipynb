## 按照格式生成qq协议的训练数据和测试数据

```python
import uuid  # 用于生成十六进制
import random
import os
```

```python
# 定义函数：生成指定范围内的随机整数

def get_int(low, high):
    n = random.randint(low, high)
    return n
```

```python
def get_two_bytes():
    res = str(uuid.uuid4())  # 形如 02680d49-e91d-4ccb-a4b9-8a05c3ed86e9
    res = res.replace('-', '')  #形如 02680d49e91d4ccba4b98a05c3ed86e9
    res = res[:4]  # 取前4个字节
    return res
```

```python
# 第一种qq报文 Request KEY

def C29():
    
    num = get_int(4, 16)
    data = ''
    for i in range(num):
        data += get_two_bytes()
        
    message = '02' + '3619' + '001d' + get_two_bytes() + '08ee307b' + data
    return message
```

```python
# 第二种qq报文 Get status of friend

def C129():
    
    num = get_int(4, 16)
    data = ''
    for i in range(num):
        data += get_two_bytes()
        
    message = '02' + '3619' + '0081' + get_two_bytes() + '08ee307b' + data
    return message
```

```python
# 第三种qq报文 Get friend online

def C39():
    
    num = get_int(4, 16)
    data = ''
    for i in range(num):
        data += get_two_bytes()
        
    message = '02' + '3619' + '0027' + get_two_bytes() + '08ee307b' + data
    return message
```

```python
# 第四种qq报文 Group name operation

def C60():
    
    num = get_int(4, 16)
    data = ''
    for i in range(num):
        data += get_two_bytes()
        
    message = '02' + '3619' + '003c' + get_two_bytes() + '08ee307b' + data
    return message
```

## 批量生成输入

```python
for i in range(400):
    
    class_id = get_int(1,4)
    print(class_id)
    if class_id == 1:
        message = C29()
    if class_id == 2:
        message = C129()
    if class_id == 3:
        message = C39()
    if class_id == 4:
        message = C60()
    
    write_file_name = 'input\\' + str(i+1) + '.txt'
    f = open(write_file_name, 'w')
    f.write(message)
    f.close()
```
